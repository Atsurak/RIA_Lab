#### Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Data URL's

```python
url1 = 'https://raw.githubusercontent.com/Atsurak/RIA_Lab/main/PUMA/g981.csv'
url2 = 'https://raw.githubusercontent.com/Atsurak/RIA_Lab/main/PUMA/g850.csv'
url3 = 'https://raw.githubusercontent.com/Atsurak/RIA_Lab/main/PUMA/g661.csv'
```

Getting Data
1. plot for g = 9.8 

```python
data = pd.read_csv(url1)
data.head()
```

```python
def lone_plot(x,y):
  plt.figure()
  plt.plot(x,y)
  plt.show()
```

```python
data.columns
```

```python
def plotter(url):
  data = pd.read_csv(url)  
  for i in range(1,7):
    y = data[f"Joint{i}: GenForce (N.m)"] # Torque for Joint i

    x = data[f"Joint{i}: JointValue (degree)"] # Torque vs Angle
    plt.figure()
    plt.subplot(1,3,1)
    plt.plot(x,y)
    x = data[f"Joint{i}: JointVel (degree/s)"] # Torque vs Angular Velocity
    plt.subplot(1,3,2)
    plt.plot(x,y)
    x = data[f"Joint{i}: JointAccl (degree/s^2)"] # Torque vs Angular Acceleration
    plt.subplot(1,3,3)
    plt.plot(x,y)
    plt.show()
```

plots for all joints at g = 9.8

```python
plotter(url1)
```

plots for all joints at g = 8.5

```python
plotter(url2)
```

```python
plotter(url3)
```

from the above plots we can see that the torque that is required at Joint 6 is very minimal and is maximum at Joint 1 so let's compare plots for different g values at Joint 1

```python
i = 1 # Joint 1
y1 = data[f"Joint{i}: GenForce (N.m)"] # Torque for Joint i
x = data[f"Joint{i}: JointValue (degree)"] # Torque vs Angle
lone_plot(x,y1)
x = data[f"Joint{i}: JointVel (degree/s)"] # Torque vs Angular Velocity
lone_plot(x,y1)
x = data[f"Joint{i}: JointAccl (degree/s^2)"] # Torque vs Angular Acceleration
lone_plot(x,y1)
```
